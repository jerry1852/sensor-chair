```python
import pandas as pd
import neurokit2 as nk
import numpy

data = pd.read_excel('D://totalsensors/data5.xlsx')
eda_signal = data.iloc[:, 1].dropna()
window_size = 100
eda_signal= eda_signal.rolling(window=window_size).mean()
eda_signal = eda_signal.dropna()
nk.signal_filter(eda_signal, highcut=0.5,method='butterworth', order=2)
signals, info = nk.eda_process(eda_signal, sampling_rate=25)
nk.eda_plot(signals, info)
```
